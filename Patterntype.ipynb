# 1. Right Triangle Star Pattern
# "*" x 1 ---> *
# "*" x 2 ---> **
# "*" x 3 ---> **&
# "*" x 4 ---> ****

```python
n = 4
for i in range(1, n+1):
    print("*" * i)        # "*" is repeated i times
```

# 2. Right Aligned Triangle (Sometimes called right triangle)

```python
n = 4
for i in range(1,n+1):
    print(" " * (n-i) + "*" * i)
```

# 3. Reverse Star Pattern
# range(start,stop,step)  n[starting value], 0[stop value , loop stops before 0], -1[Each time value decreases by 1]
# +1 - go forword  -1 - go backward

```python
n = 4
for i in range(n,0,-1):   
    print("*" * i)
```

# 4. Pyramid Pattern


```python
n = 4
for i in range(1, n+1):
    print(" " * (n-i) + "*" * (2*i - 1))

# " " * (n-i) ....> Print spaces , spaces decreases each row
# "*" * (2*i - 1) .....> Print star, star increases as odd numbers : 1,3,5,7
```

# 5. Increasing Number Pattern


```python
n = 4
for i in range(1, n+1):       # outer loop controls rows(1 to 4)
    for j in range(1, i+1):   # Inner loop , prints numbers from 1 to i
        print(j, end = " ")   # end = "" helps to print the number in same line like 123 not in vertical
    print()                   # without print(), output will be 1121231234, print() is used to move to the next line

```

```python
range(1, i+1) means
1      range(1,2)   1
2      range(1,3)   1,2
3      range(1,4)   1,2,3
4      range(1,5)   1,2,3,4
Always (start, stop-1)
```

# 6. Same Number triangle Pattern

```python
n = 4
for i in range(1, n+1):
    for j in range(1, i+1):
        print(i, end = "")
    print()
```

# 7. Floyd's Triangle

```python
n = 4
num = 1
for i in range(1,n+1):
    for j in range(1, i+1):
        print(num, end= " ")
        num = num + 1
    print()
```

# 8. Alpahbet Pattern

```python
n = 4
for i in range(1,n+1):
    for j in range(1, i+1):
        print(chr(64+j), end = " ")   # ch(65) = A, ch(66) = B, ch(67) = C, ch(68) = D 
    print()
```

# 9. Square Pattern

```python
n = 4
for i in range(1, n+1):
    for j in range(1, n+1):
        print("*", end = "")
    print()
```

# 10. Hollow Square Pattern

```python
n = 4
for i in range(1, n+1):
    for j in range(1, n+1):
        if i == 1 or i == n or j == 1 or j == n:
            print("*", end = "")
        else:
            print(" ", end = "")
    print()
```
