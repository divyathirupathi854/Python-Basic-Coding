# 6. CHECK WHETHER A NUMBER IS PALINDROME

```python
num = int(input("Enter a Number:"))
rev = 0
temp = num

while temp > 0:
    digit = temp % 10 
    rev = rev * 10 + digit         
    temp = temp // 10

if rev == num:
    print("It is a palindrome")
else:
    print("It is not a palindrome")

# or  

num = input("Enter a Number:")
rev = num[::-1]
if rev == num:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

# 7. PALINDROME USING FUNCTION FOR STRING INPUT

"return" is used to send a result back from a function
return True means "Yes"
return False means "No"

without return function keeps the answer inside itself and does not tell the outside code

If the function returns True, the "if" block executes
If the function returns False, the "else" block executes



```python
# normal method
def palindrome(s):
    rev = ""
    for ch in s:
        rev = ch + rev
    if s == rev:
        return True
    else:
        return False

s = input("Enter a String:")
if palindrome(s):
    print("palindrome")
else:
    print("Not a palindrome")

# slicing inside function (mostly avoid this method)

def palindrome(s):
    rev = s[::-1]
    if s == rev:
        return True
    else:
        return False

s = input("Enter String:")
if palindrome(s):
    print("palindrome")
else:
    print(" Not palindrome")
```

# 8. PALINDROME USING FUNCTION FOR NUMBER INPUT

# 26. ARMSTRONG NUMBER:


```python
nums = int(input("Enter a Number:"))
n = len(str(nums))
sum = 0
temp = nums
                            
while temp > 0:
    digit = temp % 10
    sum = sum + digit ** n
    temp = temp // 10

if sum == nums:
    print("It is an Armstrong number")
else:
    print("It is not an Armstrong number")
```

# 76. MERGE TWO DICTIONARIES

```python
d1 = {"a" : 1, "b" : 2}
d2 = {"c" : 3, "d" : 4}
merged = {}

for key in d1:
    merged[key] = d1[key]
for key in d2:
    merged[key] = d2[key]

print(merged)

# using update()
d1 = {"a" : 1, "b" : 2}
d2 = {"c" : 3, "d" : 4}
d1.update(d2)
print(d1)
```

# 77. REPLACE A WORD IN A STRING

```python
text = "I love watching cartoons"
old_word = "cartoons"
new_word = "movies"
words = text.split()

for i in range(len(words)):
    if words[i] == old_word:
        words[i] = new_word
result = " ".join(words)
print(result)
```

# 78. Linear search: Checking elements one by one from start to end until we find the target. PURPOSE : To find an element in a list/array


```python
arr = [10,20,30,40,50,60,70,80,90]
key = 60
found = False
for i in range(len(arr)):
    if arr[i] == key:
        print("Found key at index", i)
        found = True
        break
else:
        print("Key not found")
        
```

# 79. Binary Search: Search by dividing the array into 2 parts again and again,instead of checking one by one(like linear search), we go to the middle element first.
# Condition: Binary search works only on "SORTED ARRAY"
# Idea: check middle element:
#       if mid value = key then found
#       if smaller = go left side
#       if bigger = go right side

```python
arr = [10,20,30,40,50,60,70,80]
key = 50
low = 0
high = len(arr)-1

while low <= high:
    mid = (low+high) // 2

    if arr[mid] == key:
        print("Found key at index", mid)
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
else:
    print("key not found")

```

# 80. Bubble Sort: Compare two adjacent elements and swap if wrong order
# Purpose: To sort elements(Ascending or descending)

```python
arr = [2,6,5,3,1,4]
n = len(arr)

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
print(arr)
```

# 81. Missing Number

```python
arr = [1,2,4,5]
n = len(arr) + 1  # length of the array is 4, but one number is missing so +1

total = n * (n+1) // 2
arr_sum = sum(arr)

missing = total - arr_sum
print("Missing Number =", missing)
```

# 82. Move zeros to the end

```python
arr = [0,1,0,3,12]
position = 0       # position to place non - zero 

for i in range(len(arr)):
    if arr[i] != 0:
        arr[position], arr[i] = arr[i], arr[position]
        position = position + 1
print(arr)
```
